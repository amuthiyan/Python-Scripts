```python
from sklearn.preprocessing import LabelEncoder
```

```python
for var in var_mod:
    lb = LabelEncoder()
    full_var_data = battles[var].astype('str')
    temp = lb.fit_transform(np.array(full_var_data))
    battles[var] = lb.transform(np.array( battles[var] ).astype('str')) 
```

```python
predictors=['year','attacker_king','defender_king','attacker_1','attacker_2','attacker_3','attacker_4',\
           'defender_1','defender_2','defender_3','defender_4','attacker_outcome','battle_type','major_death',\
           'major_capture','attacker_size','defender_size','attacker_commander','defender_commander','summer',\
           'location','region','diff_size','numAttackC','numDefendC','diffCommand','attack_Fam','defend_Fam']
from sklearn.tree import DecisionTreeClassifier
from sklearn.cross_validation import cross_val_score
from sklearn.cross_validation import train_test_split
from sklearn.metrics import accuracy_score, classification_report,confusion_matrix
```
